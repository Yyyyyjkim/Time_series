# 시간 단위 데이터로 변환

```python
import re
```

```python
pv_1['Time_hour'] = [re.sub(':\d{2}$', '', i) for i in pv_1['LocalTime']]
```

```python
pv_1_hour = pv_1[['Time_hour', 'Power(MW)']].groupby('Time_hour').sum().reset_index()
```

```python
pv_1_hour.to_csv('data/PV_hour.csv', index=None)
```
